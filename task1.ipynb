```python
import pandas as pd
```

```python
df= pd.read_csv('netflix_titles.csv')
df.head()
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Fill missing 'director' and 'cast' with 'Unknown'
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
```

```python
# Drop rows where 'country' is missing
df.dropna(subset=['country'], inplace=True)
```

```python
#check for missing values
df.isnull().sum()
```

```python
# For date_added dropping
df.dropna(subset=['date_added'], inplace=True)

# For rating
df['rating'] = df['rating'].fillna('Unknown')

# For duration
df['duration'] = df['duration'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
#remove duplicates
df.drop_duplicates(inplace=True)
```

```python
#Standardize Text Values
stand_cols=  ['type', 'title', 'director', 'cast', 'country', 'rating']
# Apply lowercasing and strip spaces
for col in stand_cols:
    df[col] = df[col].astype(str).str.lower().str.strip()
```

```python
# Convert 'date_added' to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract year
df['year_added'] = df['date_added'].dt.year
```

```python
#rename column headers 
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head()
```
